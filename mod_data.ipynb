```python
import pandas as pd
df = pd.read_csv('OPV.csv')
```

```python
df
```

```python
for f in df.columns:
    print(f)
```

```python
smiles = df["SMILES"].values
```

```python
value = df["PCE_ave"].values
```

```python
new_df = pd.DataFrame({"smiles" : smiles, "value" : value})
```

```python
new_df.head()
```

```python
new_df["value"].describe()
```

```python
new_df["cat_value"] = pd.cut(
    new_df["value"],
    bins=[-float('inf'), 4, 9, float('inf')],  # Define the bins
    labels=[0, 1, 2]
)
```

```python
new_df.head()
```

```python
new_df["value"] = new_df["cat_value"]
new_df.drop(["cat_value"], axis=1, inplace=True)
```

```python
new_df.head()
```

```python
new_df.to_csv("OPV_cat.csv", index=False)
```
